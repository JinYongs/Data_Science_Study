### 앙상블 (ensemble)
- 앙상블 쓰는 이유 : 여러 개의 단일 모델들의 평균치를 내거나, 투표를 해서 다수결에 의한 결정을 하는 등 더 나은 결과를 도출해 내기 위한 목적기법 
- 배깅방법과 부스팅 방법이 있고, 배깅에는 RandomForest / 부스팅방법에는 AdaBoost, GradientBoost(GBM) , AdaBoost, XGBoost, LightGradientBoost 등이 있음.

### (1) Bagging  - Random Forest
- 샘플을 여러번 뽑아 각 모델을 학습시켜 결과를 집계
- 복원추출(뽑은거 다시 넣음)로 N개의 데이터셋을 만들고 N개의 모델을 만듬
- 다양한 데이터 셋 사용 하니까 Overfitting 방지
- Classification과 regression 모두 사용 가능
![배깅](./img/배깅.png)

### (1) - 1. 랜덤 포레스트
- 여러 개의 의사 결정 나무를 만들고 그들의 다수결로 결과를 결정하는 방법
- 성능이 좋고, 함수가 간단해서 하이퍼 파라미터 튜닝 없이 간단히 쓰기 가능

![rf](./img/rf.png)    

### (2) Boosting 
- 배깅의 변형으로 모델이 잘 예측하지 못하는 부분을 개선하기 위한 모델
- 이전 모델들이 **예측하지 못한 Error data에 가중치를 부여**하여, 다음 모델이 더 잘 예측되도록 하는 것.
- Adaboost (에이다부스트), GBM (그레디언트 부스트), XGBoost, LGBM
![부스팅](./img/부스팅.png)

### (3) 배깅 vs 부스팅
- **Bagging은 일반적인 모델을 만드는데 집중되어 있는 반면, Boosting은 맞추기 어려운 문제를 맞추는데 초점이 맞춰져 있음.**
- Boosting은 속도가 느리지만 Bagging에 비해 에러가 적고, 데이터를 사용한 걸 재사용하니 오버피팅 될 가능성이 있음.
- ==> 개별 결정 트리의 낮은 성능이 문제라면 Boosting / 오버피팅이 문제라면 Bagging

```python
# 4. 데이터 모델링 - (1) 훈련 데이터 및 검증 데이터 생성

# dacon 제출용 submission id
#Obtain the submission ID to create the submission file later
sub_id = test_prod.inst_id # 병원 ID

# 종속변수, 특징변수 분리
answer = 'OC' # 훈련 데이터 정답 라벨
not_answer = train_prod.columns.difference([answer])  # OC를 제외한 columns list 

# 훈련 데이터 / 훈련 데이터 정답 라벨 / 테스트 데이터 생성.
train_prod_X = train_prod[not_answer] # 훈련 x
train_prod_Y = train_prod[answer] # 훈련 y
test_prod_X = test_prod[not_answer] # 테스트 x

print("데이터 모델링 - (1) 훈련 데이터 및 검증 데이터 생성 완료")
```

```python
# 4. 데이터 학습 - (2) 라이브러리 import
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score

import xgboost as xgb
import seaborn as sns
import matplotlib.pyplot as plt

print('데이터 모델링 - (2) 라이브러리 import 완료')
```

```python
############ Random Forest
# 4. 데이터 모델링 - (3) Random Forest 변수 설정 및 학습 및 테스트

# 앙상블 배깅 랜덤포레스트 Bagging RandomForest
estimators = 100 # 분류기 갯수
np.random.seed(100)
RF_prod = RandomForestClassifier(n_estimators = estimators)

# 랜덤 포레스트 훈련 데이터 학습
RF_prod_model = RF_prod.fit(train_prod_X, train_prod_Y) 

# 랜덤 포레스트 테스트 데이터 테스트 
RF_prod_prediction = RF_prod.predict_proba(test_prod_X)[:,1] 

# 제출용 파일 생성 (sub_id 는 test Id )
sub_RF = pd.DataFrame({'inst_id' : sub_id , 'OC' : RF_prod_prediction })
sub_RF = sub_RF[['inst_id', 'OC']]

print('모델 정보\n', RF_prod_model)
print("데이터 모델링 - (3) Random Forest 변수 설정 및 학습 및 테스트 완료")

print('\n테스트 결과\n')
sub_RF
```

```python
############ GBM: Gradient Boosting
# 4 데이터 모델링 - (4) GBM 변수 설정 및 학습 및 테스트
# 부스팅 방법 중 GBM 
GBM_prod = GradientBoostingClassifier(n_estimators = estimators)

# GBM 훈련 데이터 학습
GBM_prod_model = GBM_prod.fit(train_prod_X, train_prod_Y) 

# GBM 테스트 데이터 테스트
GBM_prod_prediction = GBM_prod.predict_proba(test_prod_X)[:,1]

# 제출용 파일 생성
sub_GBM = pd.DataFrame({'inst_id' : sub_id , 'OC' : GBM_prod_prediction })
sub_GBM = sub_GBM[['inst_id', 'OC']]

print('모델 정보\n', GBM_prod_model)
print("데이터 모델링 - (4) GBM 변수 설정 및 학습 및 테스트 완료")
print('\n GBM 테스트 결과\n')
sub_GBM
```

```python
############ XGBOOST
# GBM 보다 성능이 좀더 좋음. -> 더 좋은건 lgbm
# 4. 데이터 모델링 - (5) XGBOOST 변수 설정 및 학습 및 테스트
dtrain_prod = xgb.DMatrix(data = train_prod_X, label = train_prod_Y)
dtest_prod = xgb.DMatrix(data = test_prod_X)

# Custom error function for the XGB model
threshold = 0.5
def eval_error(preds, dtrain):
    labels = dtrain.get_label()
    preds = (preds > threshold ).astype('float')
    return "accuracy", accuracy_score(labels, preds)
    
# xgboost - reg:linear(linear-regression / binary:logistic classification) 
param = {'objective' : 'binary:logistic',  # 목적함수
         'max_depth' : 6, # leaf node 2^6 = 64 
         'eta': 0.3, # learning late
         'colsample_bytree' : 1,
         'subsample' : 1,
         # 'silent' : 0 -> not be used error
         # eval_metrics : 모델의 평가 함수 조정하는 함수
         }

nrounds = 100 # epoch / 반복 라운드 결정

np.random.seed(100)

# xgb_model 학습 
xgb_model = xgb.train(param, 
                      dtrain_prod, 
                      num_boost_round = nrounds , # = epoch 반복횟수
                      feval = eval_error, 
                      #maximize = True,
                      #early_stopping_rounds = 10,
                      )

XGB_prediction = xgb_model.predict(dtest_prod)

sub_XGB= pd.DataFrame({'inst_id' : sub_id , 'OC' : XGB_prediction })
sub_XGB= sub_XGB[['inst_id', 'OC']]

print('\n XGBoost 테스트 결과\n')
sub_XGB

#print("데이터 모델링 - (5) XGBOOST 변수 설정 및 학습 및 테스트 완료")
```

```python
# 4. 데이터 모델링 - (6) 앙상블 진행. 각 기법 훈련 결과 csv로 통합 저장
#Forming the ensemble dataset of the 3 models
ensemble = pd.DataFrame()
ensemble['inst_id'] = sub_XGB['inst_id']
ensemble['RandomForest'] = sub_RF['OC']
ensemble['GBM'] = sub_GBM['OC']
ensemble['XGB'] = sub_XGB['OC']

# Taking the average of all 3 models
# 모델 3개 합치고 0.7미만인 것들은 폐업으로 예상
ensemble['ens'] = (ensemble['XGB'] + ensemble['GBM'] + ensemble['RandomForest'])/3
ensemble['OC'] = (ensemble['ens'] > 0.7).astype('int') #0.7 as the threshold and above that the hospital is closed.

ensemble.to_csv('hospital_submission.csv', index=False)
#ensemble.sort_values('ens', ascending=False)
ensemble
```

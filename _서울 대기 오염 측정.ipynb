```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=-1, strategy='mean')
df_imputed = pd.DataFrame(imp.fit_transform(df[["SO2","NO2","O3","CO","PM10","PM2.5"]]))
df_imputed.columns = df[["SO2","NO2","O3","CO","PM10","PM2.5"]].columns
df_imputed.index = df.index
remain_df = df[df.columns.difference(["SO2","NO2","O3","CO","PM10","PM2.5"])]
df_train = pd.concat([remain_df, df_imputed], axis=1)
df_train.head()
```

```python
last_entry = df.groupby('Station code').max()
last_entry
```

```python
def get_colors(data, df_item, item):
    item_row = df_item.loc[df_item['Item name'] == item]
    if (data > df_item.iloc[0]['Very bad(Red)']):
        return 'red'
    elif (data > df_item.iloc[0]['Bad(Yellow)']):
        return 'yellow'
    elif (data > df_item.iloc[0]['Normal(Green)']):
        return 'green'
    else:
        return 'blue'
```

```python
last_entry['SO2 Color'] = last_entry['SO2'].apply(get_colors, args =(df_item, 'SO2' )) 
last_entry['NO2 Color'] = last_entry['NO2'].apply(get_colors, args =(df_item, 'NO2' )) 
last_entry['O3 Color'] = last_entry['O3'].apply(get_colors, args =(df_item, 'O3' )) 
last_entry['CO Color'] = last_entry['CO'].apply(get_colors, args =(df_item, 'CO' )) 
last_entry['PM10 Color'] = last_entry['PM10'].apply(get_colors, args =(df_item, 'PM10' )) 
last_entry['PM2.5 Color'] = last_entry['PM2.5'].apply(get_colors, args =(df_item, 'PM2.5' )) 
last_entry
```
